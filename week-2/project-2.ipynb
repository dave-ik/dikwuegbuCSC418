```python
def main():
        users= {
            'AJ': (1, 'C:/Users/2122/Desktop/dikwuegbuCSC418/dikwuegbuCSC418/week-2/class/Aj.jpeg'),
           'Anderson': (2, 'C:/Users/2122/Desktop/dikwuegbuCSC418/dikwuegbuCSC418/week-2/class/Anderson.jpeg'),
            'Bruno': (3, 'C:/Users/2122/Desktop/dikwuegbuCSC418/dikwuegbuCSC418/week-2/class/Bruno.jpeg'),
            'Chinwendu': (4, 'C:/Users/2122/Desktop/dikwuegbuCSC418/dikwuegbuCSC418/week-2/class/Chinwendu.jpeg'),
            'Fatima': (5, 'C:/Users/2122/Desktop/dikwuegbuCSC418/dikwuegbuCSC418/week-2/class/Fatima.jpeg'),
            'Keisha': (6, 'C:/Users/2122/Desktop/dikwuegbuCSC418/dikwuegbuCSC418/week-2/class/Keisha.jpeg'),
            'Lotanna': (7, 'C:/Users/2122/Desktop/dikwuegbuCSC418/dikwuegbuCSC418/week-2/class/Lotanna.jpeg'),
            'Ogb': (8, 'C:/Users/2122/Desktop/dikwuegbuCSC418/dikwuegbuCSC418/week-2/class/Ogb.jpeg'),
            'Sope': (9, 'C:/Users/2122/Desktop/dikwuegbuCSC418/dikwuegbuCSC418/week-2/class/Sope.jpeg'),
            'Zion': (10, 'C:/Users/2122/Desktop/dikwuegbuCSC418/dikwuegbuCSC418/week-2/class/Zion.jpeg')
    }
        username= input("Enter username: ")
        password= int(input("Enter passkey(matric: )"))

        if username in users:
            matric, path = users[username]

            correct_password = matric

            if password == correct_password:
                
                try:
                    
                    Option= input(f"Select action: 1 for Contrast, 2 for Sharpen, 3 for Noise, 4 for Scaling, 5 for Inverse, 6 for Arithmetic Opss: ")

                    match Option:
                        case '1':
                            image = cv2.imread(path)
                            plt.subplot(1, 2, 1)
                            plt.title("Original")
                            plt.imshow(image)

                            brightness  = 5
                            contrast = 1.5
                            image2 = cv2.addWeighted(image, contrast, np.zeros(image.shape, image.dtype), 0, brightness)

                            cv2.imwrite(f'img/{username}_Contrast.jpg', image2)

                            plt.subplot(1,2,2)
                            plt.title(f"{username} contrast")
                            plt.imshow(image2)
                            plt.show()

                        case '2':
                            image = cv2.imread(path)
                            plt.subplot(1,2,1)
                            plt.title(f'original {username}')
                            plt.imshow(image)
                            kernel = np.array([[0,-1,0],[-1,5,-1],[0,-1,0]])
                            sharpened_image = cv2.filter2D(image, -1, kernel)
                            cv2.imwrite(f'img/{username}_Sharpened.jpg', sharpened_image)
                            plt.subplot(1,2,2)
                            plt.title(f"{username} Sharpen")
                            plt.imshow(sharpened_image)
                            plt.show()

                        case '3':
                            image = cv2.imread(path)
                            plt.subplot(1,2,1)
                            plt.title(f'original {username}')
                            plt.imshow(image)
                            filtered_image = cv2.medianBlur(image, 15)
                            cv2.imwrite(f'img/{username}_Noiseless.jpg', filtered_image)
                            plt.subplot(1,2,2)
                            plt.title(f"{username} Noise")
                            plt.imshow(filtered_image)
                            plt.show()

                        case '4':
                            image = cv2.imread(path)
                            plt.subplot(1,2,1)
                            plt.title(f'original {username}')
                            plt.imshow(image)
                            scaled_image = cv2.resize(image, None, fx=2, fy=2)
                            cv2.imwrite(f'img/{username}_Scaled.jpg', scaled_image)
                            plt.subplot(1,2,2)
                            plt.title(f"{username} Scaled")
                            plt.imshow(scaled_image)
                            plt.show()

                        case '5':
                            image = cv2.imread(path)
                            plt.subplot(1,2,1)
                            plt.title("Original")
                            plt.imshow(image)

                            inverse_image = 255-image

                            cv2.imwrite(f'img/{username}_inverse.jpg',inverse_image)

                            plt.subplot(1,2,2)
                            plt.title(f"{username}, Inversed")
                            plt.imshow(inverse_image)
                            plt.show()

                        case '6':
                            image = cv2.imread(path)

                
                except Exception as e:
                    print('Error:', e)                       
            else:
                print("Password Not seen") 
        else:
            print("Authentication failed. User not found.")
```
