```python
import cv2
import matplotlib.pyplot as plt
import numpy as np
```

```python
image = cv2.imread('C:/Users/2122/Downloads/city 3.jpg')
```

```python
plt.subplot(1, 2, 1)
plt.title("Original")
plt.imshow(image)
```

```python
brightness = 5
contrast = 1.5
image2 = cv2.addWeighted(image, contrast, np.zeros(image.shape, image.dtype), 0, brightness)
```

```python
cv2.imwrite('C:/Users/2122/Downloads/contrast_image.jpg', image2)
```

```python
plt.subplot(1, 2, 2)
plt.title("Brightness & Contrast")
plt.imshow(image2)
plt.show()
```
