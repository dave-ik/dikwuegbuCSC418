```python
import cv2
import matplotlib.pyplot as plt
import numpy as np
```

```python
image = cv2.imread('C:/Users/2122/Desktop/dikwuegbuCSC418/dikwuegbuCSC418/week-2/sst-foyer.jpg')
```

```python
plt.subplot(1, 2, 1)
plt.title("Original")
plt.imshow(image)
```

```python
kernel = np.array([[0, -1, 0], [-1, 5, -1], [0, -1, 0]])
```

```python
sharpened_image = cv2.filter2D(image, -1, kernel)
```

```python
cv2.imwrite('C:/Users/2122/Desktop/dikwuegbuCSC418/dikwuegbuCSC418/week-2/sharpened_image.jpg', sharpened_image)
```

```python
plt.subplot(1, 2, 2)
plt.title("Sharpening")
plt.imshow(sharpened_image)
plt.show()
```
