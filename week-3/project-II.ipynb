```python
import tkinter as tk
from tkinter import messagebox
import cv2
import numpy as np
```

```python

images = {
    'Traditional': {
        'Benin King Bust': 'C:/Users/2122/Desktop/dikwuegbuCSC418/week-3/Traditional/Benin King Bust.jpg',
        'Benin Queen': 'C:/Users/2122/Desktop/dikwuegbuCSC418/week-3/Traditional/Benin Queen Bust.jpg'
    },
    'Modern': {
        'Iya Abikun': 'C:/Users/2122/Desktop/dikwuegbuCSC418/week-3/Modern/Iya Abikun.jpg',
        'Red Sky': 'C:/Users/2122/Desktop/dikwuegbuCSC418/week-3/Modern/Red Sky.jpg' 
    },
    'Contemporary': {
        'Eti Osa': 'C:/Users/2122/Desktop/dikwuegbuCSC418/week-3/Contemporary/Eti Osa.jpg',
        'Home Bond': 'C:/Users/2122/Desktop/dikwuegbuCSC418/week-3/Contemporary/Home Bond.jpg' 
    }
}

def login():
    username = username_entry.get()
    password = password_entry.get()
    login_window.destroy() 
    show_category_selection()

def show_category_selection():
    category_window = tk.Tk()
    category_window.title("Select Category")

   
    category_label = tk.Label(category_window, text="Category:")
    category_label.grid(row=0, column=0, padx=10, pady=5)
    category_entry = tk.Entry(category_window)
    category_entry.grid(row=0, column=1, padx=10, pady=5)

   
    submit_button = tk.Button(category_window, text="Submit", command=lambda: show_selected_category(category_entry))
    submit_button.grid(row=1, column=0, columnspan=2, padx=10, pady=10)

def show_selected_category(category_entry):
    category = category_entry.get()
    if category == "Traditional":
        traditional_viewer()
    elif category == "Modern":
        modern_viewer()
    elif category == "Contemporary":
        contemporary_viewer()
    else:
        messagebox.showerror("Invalid Category", "Category does not exist")

def traditional_viewer():
    root = tk.Tk()
    root.title("Traditional Art")
    Lb = tk.Listbox(root)
    for artwork in images['Traditional']:
        Lb.insert(tk.END, artwork)
    Lb.pack()

   
    def on_select(event):
        index = Lb.curselection()
        if index:
            selected_artwork = Lb.get(index)
            enhancement_window(selected_artwork, images['Traditional'][selected_artwork])
    
    
    Lb.bind("<<ListboxSelect>>", on_select)

def modern_viewer():
    root = tk.Tk()
    root.title("Modern Art")
    Lb = tk.Listbox(root)
    for artwork in images['Modern']:
        Lb.insert(tk.END, artwork)
    Lb.pack()

    
    def on_select(event):
        index = Lb.curselection()
        if index:
            selected_artwork = Lb.get(index)
            enhancement_window(selected_artwork, images['Modern'][selected_artwork])
    
    Lb.bind("<<ListboxSelect>>", on_select)

def contemporary_viewer():
    root = tk.Tk()
    root.title("Contemporary Art")
    Lb = tk.Listbox(root)
    for artwork in images['Contemporary']:
        Lb.insert(tk.END, artwork)
    Lb.pack()

    def on_select(event):
        index = Lb.curselection()
        if index:
            selected_artwork = Lb.get(index)
            enhancement_window(selected_artwork, images['Contemporary'][selected_artwork])
    
  
    Lb.bind("<<ListboxSelect>>", on_select)

def enhancement_window(artwork_name, artwork_path):
    root = tk.Tk()
    root.title("Select Enhancement")
    
    enhancements = ["Contrast", "Sharpness", "Noise", "Scaling", "Transformation", "Arithmetic", "Colour"]

    selected_enhancement = tk.StringVar(root)
    
    enhancement_label = tk.Label(root, text="Select Enhancement:")
    enhancement_label.pack(pady=5)
    
    enhancement_menu = tk.OptionMenu(root, selected_enhancement, *enhancements)
    enhancement_menu.pack(pady=5)
    
    apply_button = tk.Button(root, text="Apply Enhancement", command=lambda: process_enhancement(artwork_name, artwork_path, selected_enhancement.get()))
    apply_button.pack(pady=5)

def process_enhancement(artwork_name, artwork_path, selected_enhancement):
    if selected_enhancement == "Arithmetic":
        arithmetic_window(artwork_name, artwork_path)
    elif selected_enhancement == "Colour":
        colour_window(artwork_name, artwork_path)
    else:
        apply_enhancement(artwork_path, selected_enhancement)

def arithmetic_window(artwork_name, artwork_path):
    root = tk.Tk()
    root.title("Select Arithmetic Operation")

   
    operations = ["Addition", "Subtraction"]

    selected_operation = tk.StringVar(root)

    operation_label = tk.Label(root, text="Select Arithmetic Operation:")
    operation_label.pack(pady=5)

    operation_menu = tk.OptionMenu(root, selected_operation, *operations)
    operation_menu.pack(pady=5)
   
    apply_button = tk.Button(root, text="Apply", command=lambda: apply_arithmetic_window(artwork_path, selected_operation.get()))
    apply_button.pack(pady=5)

def apply_arithmetic_window(artwork_path, selected_operation):
    enhanced_image = apply_arithmetic(artwork_path, selected_operation)
    display_opencv_image(enhanced_image, "Arithmetic Enhancement")

def colour_window(artwork_name, artwork_path):
    root = tk.Tk()
    root.title("Select Colour")

   
    colours = ["Red", "Blue", "Green"]

    selected_colour = tk.StringVar(root)

    colour_label = tk.Label(root, text="Select Colour:")
    colour_label.pack(pady=5)
 
    colour_menu = tk.OptionMenu(root, selected_colour, *colours)
    colour_menu.pack(pady=5)
 
    apply_button = tk.Button(root, text="Apply", command=lambda: apply_colour_window(artwork_path, selected_colour.get()))
    apply_button.pack(pady=5)

def apply_colour_window(artwork_path, selected_colour):
    enhanced_image = apply_colour(artwork_path, selected_colour)
    display_opencv_image(enhanced_image, "Colour Enhancment")

def apply_enhancement(image_path, enhancement):
   
    if enhancement == "Contrast":
        enhanced_image = apply_contrast(image_path)
    elif enhancement == "Sharpness":
        enhanced_image = apply_sharpness(image_path)
    elif enhancement == "Noise":
        enhanced_image = apply_noise(image_path)
    elif enhancement == "Scaling":
        enhanced_image = apply_scaled(image_path)
    elif enhancement == "Transformation":
        enhanced_image = apply_transformation(image_path)
    elif enhancement == "Arithmetic":
        enhanced_image = apply_arithmetic(image_path)
    elif enhancement == "Colour":
        enhanced_image = apply_colour(image_path)
    else:
        messagebox.showerror("Invalid Enhancement", "Invalid enhancement option")
        return
    
    display_opencv_image(enhanced_image, "Enhanced Image")

def apply_contrast(image_path):
    image = cv2.imread(image_path)
    contrast = 1
    brightness = 0  
    image2 = cv2.addWeighted(image, contrast, np.zeros(image.shape,image.dtype), 0, brightness)
    display_opencv_image(image2, "Contrast")

def apply_sharpness(image_path):
    image = cv2.imread(image_path)
    kernel = np.array([[0,-1,0],[-1,5,-1],[0,-1,0]])
    sharpened_image = cv2.filter2D(image, -1, kernel)
    display_opencv_image(sharpened_image, "Sharpened")

def apply_noise(image_path):
    image = cv2.imread(image_path)
    filtered_image = cv2.medianBlur(image, 15)
    display_opencv_image(filtered_image, "Noisy")

def apply_scaled(image_path):
    image = cv2.imread(image_path)
    scaled_image = cv2.resize(image, None, fx=2, fy=2)
    display_opencv_image(scaled_image, "Scaled")

def apply_transformation(image_path):
    image = cv2.imread(image_path)
    inverse_image = 255 - image
    display_opencv_image(inverse_image, "Transformed")

def apply_arithmetic(image_path, operation):
    image1 = cv2.imread(image_path)
    image2 = cv2.imread("/Users/HP1030G2/Documents/CSC418/kaobiiwelCSC418/img/dog.jpeg")
    image1 = cv2.resize(image1, (500,400))
    image2 = cv2.resize(image2, (500,400))
    
    if operation == 'Addition':
        addImage = cv2.addWeighted(image1, 0.5, image2, 0.6, 0)
        return addImage

    elif operation == 'Subtraction':
        sub = cv2.subtract(image1, image2)
        return sub

    else:
        print("Invalid Operation")

def apply_colour(image_path, selection):
    image = cv2.imread(image_path)
    B, G, R = cv2.split(image)
    if selection == 'Red':
        return R
    elif selection == 'Blue':
        return B
    elif selection == 'Green':
        return G
    else:
        print("Bad selection")


def display_opencv_image(image, title, width=800, height=600):
    resized_image = cv2.resize(image, (width, height))
    cv2.imshow(title, resized_image)
    cv2.waitKey(0)
    cv2.destroyAllWindows()


login_window = tk.Tk()
login_window.title("Login")


username_label = tk.Label(login_window, text="Username:")
username_label.grid(row=0, column=0, padx=10, pady=5)
username_entry = tk.Entry(login_window)
username_entry.grid(row=0, column=1, padx=10, pady=5)


password_label = tk.Label(login_window, text="Password:")
password_label.grid(row=1, column=0, padx=10, pady=5)
password_entry = tk.Entry(login_window, show="*")
password_entry.grid(row=1, column=1, padx=10, pady=5)


login_button = tk.Button(login_window, text="Login", command=login)
login_button.grid(row=2, column=0, columnspan=2, padx=10, pady=10)


login_window.mainloop()
```
