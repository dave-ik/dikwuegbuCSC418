```python
import tkinter as tk
from tkinter import messagebox
import cv2
```

```python
images = {
    'Traditional': {
        'Benin King Bust': 'C:/Users/2122/Desktop/dikwuegbuCSC418/week-3/Traditional/Benin King Bust.jpg',
        'Benin Queen Bust': 'C:/Users/2122/Desktop/dikwuegbuCSC418/week-3/Traditional/Benin Queen Bust.jpg' 
    },
    'Modern': {
        'Iya Abikun': 'C:/Users/2122/Desktop/dikwuegbuCSC418/week-3/Modern/Iya Abikun.jpg',
        'Red Sky': 'C:/Users/2122/Desktop/dikwuegbuCSC418/week-3/Modern/Red Sky.jpg'
    },
    'Contemporary': {
        'Eti Osa': 'C:/Users/2122/Desktop/dikwuegbuCSC418/week-3/Contemporary/Eti Osa.jpg',
        'Home Bond': 'C:/Users/2122/Desktop/dikwuegbuCSC418/week-3/Contemporary/Home Bond.jpg'
    }
}

# Authenticate user
def authenticate(username, password):
    # For simplicity, i'm using hardcoded credentials
    if username == "user" and password == "password":
        return True
    else:
        return False

# Login button clicked
def login_clicked():
    username = username_entry.get()
    password = password_entry.get()
    if authenticate(username, password):
        messagebox.showinfo("Login Successful", "Welcome!")
        show_main_interface()
    else:
        messagebox.showerror("Login Failed", "Invalid username or password")

# Displays main interface after successful login
def show_main_interface():
    login_frame.pack_forget()
    main_frame.pack()

# Image selection
def select_image(category, artwork):
    image_path = images[category][artwork]
    image = cv2.imread(image_path)

    max_width = 600
    max_height = 400
    
    height, width, _ = image.shape
    if width > max_width or height > max_height:
        ratio = min(max_width / width, max_height / height)
        image = cv2.resize(image, (int(width * ratio), int(height * ratio)))

    # Image processing / Display the image
    cv2.imshow('Image', image)
    cv2.waitKey(0)
    cv2.destroyAllWindows()

# Creating the main window
root = tk.Tk()
root.title("Art Museum Image Viewer")

# Login frame
login_frame = tk.Frame(root)
login_frame.pack(padx=20, pady=20)
tk.Label(login_frame, text="Username:").grid(row=0, column=0)
tk.Label(login_frame, text="Password:").grid(row=1, column=0)
username_entry = tk.Entry(login_frame)
username_entry.grid(row=0, column=1)
password_entry = tk.Entry(login_frame, show="*")
password_entry.grid(row=1, column=1)
login_button = tk.Button(login_frame, text="Login", command=login_clicked)
login_button.grid(row=2, columnspan=2)

# Main interface frame
main_frame = tk.Frame(root)

# Image gallery
for category, artworks in images.items():
    tk.Label(main_frame, text=category, font=("Helvetica", 16, "bold")).pack()
    for artwork in artworks:
        tk.Button(main_frame, text=artwork, command=lambda cat=category, art=artwork: select_image(cat, art)).pack()

root.mainloop()
```
