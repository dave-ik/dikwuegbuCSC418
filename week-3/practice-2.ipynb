```python
import numpy as np
import matplotlib.pyplot as plt
import cv2 as cv
```

```python
img = cv.imread('C:/Users/2122/Desktop/dikwuegbuCSC418/week-3/sst.jpg', 0)
```

```python
# Plotting the original image
plt.subplot(1, 2, 1)
plt.title("Original")
plt.imshow(img)
```

```python
rows, cols = img.shape
M = np.float32([[1, 0, 0], [0, -1, rows], [0, 0, 1]])
reflected_img = cv.warpPerspective(img, M,(int(cols), int(rows)))
```

```python
# Plotting the reflected image
plt.subplot(1, 2, 2)
plt.title("Reflected Image")
plt.imshow(reflected_img)
```

```python
cv.imwrite('reflection_out.jpg', reflected_img)
cv.waitKey(0)
cv.destroyAllWindows()
```
