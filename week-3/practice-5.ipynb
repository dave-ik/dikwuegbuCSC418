```python
import numpy as np
import matplotlib.pyplot as plt
import cv2 as cv
```

```python
img = cv.imread('C:/Users/2122/Desktop/dikwuegbuCSC418/week-3/sst.jpg', 0)
rows, cols = img.shape
```

```python
plt.subplot(1, 2, 1)
plt.title("Original")
plt.imshow(img)
```

```python
M = np.float32([[1, 0.5, 0], [0, 1, 0], [0, 0, 1]])
sheared_image = cv.warpPerspective(img, M, (int(cols*1.5), int(rows*1.5)))
```

```python
plt.subplot(1, 2, 2)
plt.title("Sheared Image")
plt.imshow(sheared_image)
```

```python
cv.imshow('img', sheared_image)
cv.waitKey(0)
cv.destroyAllWindows()
```
